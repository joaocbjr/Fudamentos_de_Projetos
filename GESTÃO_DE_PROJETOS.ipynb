#Fundamentos de Gestão de Projetos#


##Conhecendo os elementos de um projeto##
**Projeto** é uma atividade temporária com um objetivo único e um orçamento.
* Temporário
* Objetivo único
* Orçamento



##Respostas fundamentais:##

> Qual problema você está resolvendo?

> Como irá soluciona-lo?

> Qual é o seu plano?

> Como saber se terminou?

> Até que ponto o projeto teve exito?

* O que deu certo?
* O que não deu certo?
* Por quê?
* Como poderiamos ter feito melhor?



##O que é preciso pra ser gerente de projeto##
Habilidades técnicas
* Criar um plano de projeto
* Criar e ajustar um cronograma de projeto
* Medir o desempenho

Conhecimento empresarial
> É preciso entender a atividade da sua organização e o que ela acha importante.

Solução de problemas
> Os projetos não saem como planejados, é seu trabalho descobriri como cumprir os objetivos do projeto, o cronograma e o orçamento, apezar dos problemas que surgirem.

Habilidades interpessoais
> Os projetos normalmente usam pessoas de grupos, departamentos e pessoas de outras empresas, como lider da equipe é necessário trabalhar com todos.

Forte Liderança.
* Inspirar
* Orientar
* Responsabilizar
* Motivar





## Entendendo o ciclo de  vida do projeto##
05 grupos de processos

---
[ INÍCIO ]   -   [ PLANEJAMENTO ]   -   [ EXECUÇÃO ]   -   [ MONITORAMENTO E CONTROLE ]   -   [ ENCERRAMENTO ]

---


***Início***
* Definir o projeto
* Desenvolver o escopo inicial
* Estimar o custo
* Planejar recursos
* Indentificar os interessados no projeto

***Planejamento***
* O que faremos?
* Como faremos?
* Como saberemos que chegou ao fim?

***Execução***
* Iniciar o projeto
* Montar a equipe do projeto
* Desenvolver e gerir a equipe de projeto
* Explicar as diretrizes do projeto
* Executar o plano

***Monitoramento e controle***
* Registar o desempenho
* Controlar o cronograma

***Encerramento***
* Garantir a aceitação do projeto
* Documentar o desempenho
* Reunir experiências adiquiridas
* Fechar contratos
* Liberar funcionários




##Gestão de projeto tradicional x iterativa

### Gestão de projetos tradicional (em cascata)

Início -> Planejamento -> Execução -> Monitoramento e controle -> Encerramento

* Objetivo e solução definidos
* Clareza no escopo e nas entregas

### Gestão de projetos Ágeis e interativos

* Maior envolvimento do cliente
* Equipes pequenas e independentes

> **Início e planejamento**
>* Define o objetivo geral e um plano pra atingir a meta
>* Cria o plano geral para cada interação

> **Execução**
>* Realiza uma iteração por vez
>* Garante a harmonia em equipes menores

>**Monitoramento e controle**
>* Monitora e controla de perto
>* Comunica-se com rapidez e frequência

>**Encerramento** -
cada iteração tem seu proprio processo de encerramento por aceitar entregas específicas.
